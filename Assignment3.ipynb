In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

```python
###############################################################
###                                                         ###
###   The code in this cell calculates the bar statistics   ###
###                                                         ###
###############################################################

# Magic command to set interactive backend in the notebook
%matplotlib notebook

# Import the required modules
import matplotlib.pyplot as plt

# Initial y-axis value of interest
y_val = 42000

# Create an empty DataFrame
df_stats = pd.DataFrame()

# Add the mean for each year
df_stats['mean'] = df.mean(axis=1)

# Standard deviation of the samples
df_stats['std'] = df.std(axis=1)

# Number of samples
df_stats['samples'] = df.shape[1]

# y error
# see https://www.coursera.org/learn/python-plotting/discussions/weeks/3/threads/MYBNJGIQEee2fg6zpf5eMg
df_stats['yerr'] = df_stats['std'] / np.sqrt(df_stats['samples'])

# Distance from the mean in terms of yerr
df_stats['z'] = (y_val - df_stats['mean']) / df_stats['yerr']

# Create the color mapper
# See https://www.coursera.org/learn/python-plotting/discussions/weeks/3/threads/K88WKiqtEeeqRw77WotmkA
# And https://pyhogs.github.io/colormap-examples.html
normalizer = plt.Normalize(vmin=-1.96, vmax=1.96, clip=True)
sm = plt.cm.ScalarMappable(norm=normalizer, cmap=plt.cm.RdBu)

# Create the color column
df_stats['color'] = df_stats['z'].apply(sm.to_rgba)

# Display the DataFrame
df_stats
```

```python
#######################################################################
###                                                                 ###
###   The code in this creates the function that updates the plot   ###
###                                                                 ###
#######################################################################

def update_bar_plot(event):

    # Get the new y_val from the event and round it 
    new_y_val = int(event.ydata)

    # Calculade the updated z and colors
    z = (new_y_val - df_stats['mean']) / df_stats['yerr']
    colors = z.apply(sm.to_rgba).tolist()

    # Updathe the horizontal line position
    hline.set_ydata(new_y_val)

    # Update the text
    text.set_text(new_y_val)
    text.set_y(new_y_val-600)

    # Update the colors of the bars and keep the black border
    for bar, color in zip(bars, colors):
        bar.set_color(color)
        bar.set_edgecolor('k')
        
```

```python
######################################################
###                                                ###
###   The code in this cell creates the bar plot   ###
###                                                ###
######################################################

# Import the required modules
import matplotlib as mpl

# Create the figures and the axis
# See https://stackoverflow.com/questions/44002453/matplotlib-reduce-axes-width-in-subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(7,4), gridspec_kw={"width_ratios" : [25,1]})

# Plot the bars including the errors with caps
bars = ax1.bar(x = df_stats.index, 
               height = df_stats['mean'].values, 
               yerr = 1.96 * df_stats['yerr'].values,
               capsize = 8, 
               color = df_stats['color'].values,
               edgecolor = 'k',
               alpha = 0.85)

# Draw the line on the chosen y value
hline = ax1.axhline(y_val, color='k', alpha=0.6, linestyle='--')

# Display the chosen value
text = ax1.text(1995.62, y_val-600, str(y_val), fontsize=10)

# Adjust the x ticks labels
ax1.set_xticks([1992, 1993, 1994, 1995])

# Remove the x axis ticks
ax1.xaxis.set_ticks_position('none') 

# Remove the top and right spines
ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)

# Set the x and y labels
ax1.set_xlabel("Year", fontsize=14)
ax1.set_ylabel("Y value", fontsize=14)

# Include a title
ax1.set_title("Click to update the plot", fontsize=18)

# Link the event
# See https://www.coursera.org/learn/python-plotting/lecture/W2YBx/interactivity
fig.canvas.mpl_connect('button_press_event', update_bar_plot)

# Create the colorbar
# See https://www.coursera.org/learn/python-plotting/discussions/weeks/3/threads/Q0YnGQ_zEeeijwrAcoNCCA/replies/cS7qOg_6EeeI7RJK6Ns8Rg
cbar = mpl.colorbar.ColorbarBase(ax2, cmap=plt.cm.RdBu, norm=normalizer, orientation='vertical')

# Adjust the ticks in the color bar
cbar.set_ticks([-1.96, -1.5, -1.0, -0.5, 0.0, 0.5, 1.0, 1.5, 1.96])

# Display the figure
plt.show()

```
